# Model Experiments — NLP Sentiment Analyzer

Tracks training runs, hyperparameter tuning, and ablation studies.

**Goal**: Achieve ≥94% accuracy on the test set.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Load Training History

```python
with open('../reports/training_history.json', 'r') as f:
    history = json.load(f)

epochs = [e['epoch'] for e in history['history']]
train_loss = [e['train_loss'] for e in history['history']]
val_loss = [e.get('val_loss', 0) for e in history['history']]
train_acc = [e['train_accuracy'] for e in history['history']]
val_acc = [e.get('val_accuracy', 0) for e in history['history']]

print(f'Best validation accuracy: {history["best_val_accuracy"]:.4f}')
print(f'Total training time: {history["total_time_s"]:.1f}s')
```

## 2. Loss Curves

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Loss
axes[0].plot(epochs, train_loss, 'o-', label='Train Loss', color='#14f0c4')
axes[0].plot(epochs, val_loss, 'o-', label='Val Loss', color='#f97316')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_title('Training & Validation Loss')
axes[0].legend()
axes[0].grid(True)

# Accuracy
axes[1].plot(epochs, train_acc, 'o-', label='Train Acc', color='#14f0c4')
axes[1].plot(epochs, val_acc, 'o-', label='Val Acc', color='#f97316')
axes[1].axhline(y=0.94, color='green', linestyle='--', label='Target (94%)')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('Training & Validation Accuracy')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

## 3. Final Test Metrics

```python
with open('../reports/training_metrics.json', 'r') as f:
    metrics = json.load(f)

print(f"Test Accuracy: {metrics['accuracy']*100:.2f}%\n")
print(f"{'Class':<12} {'Precision':>10} {'Recall':>10} {'F1':>10}")
print('-' * 44)
for label in ['negative', 'neutral', 'positive']:
    m = metrics.get(label, {})
    print(f"{label:<12} {m.get('precision', 0):>10.4f} {m.get('recall', 0):>10.4f} {m.get('f1-score', 0):>10.4f}")
```
